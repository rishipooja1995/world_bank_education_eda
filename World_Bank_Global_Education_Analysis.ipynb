## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# ── STEP 1: Filter only real countries (remove regional aggregates) ──
# Remove the null values in the Region column only showing the 3 specific columns except the years columns
real_countries = country[country['Region'].notna()][['Country Code', 'Region', 'Income Group']]
print("Real countries after removing aggregates:", len(real_countries))

# ── STEP 2: Filter only the 4 indicators we need ──
# Storing the values of 4 indicators that are needed
indicators_needed = [
    'SE.ADT.LITR.ZS',       # Adult literacy rate, both sexes
    'SE.ADT.LITR.FE.ZS',    # Adult literacy rate, female
    'SE.ADT.LITR.MA.ZS',    # Adult literacy rate, male
    'SE.XPD.TOTL.GD.ZS'     # Govt expenditure on education % of GDP
]

filtered_data = data[data['Indicator Code'].isin(indicators_needed)] # to check whether the column have those indicator values
print("Rows after indicator filtering:", len(filtered_data)) # No count the number of rows which return the True value

# ── STEP 3: Keep only year columns 2000–2015 + identifier columns ──
# List comprehension to find the values of year in string from 2000 to 2016.
# Why this range? --> Because we find that within this range only we found the minimum number of missing values
year_cols = [str(y) for y in range(2000, 2016)]
keep_cols = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'] + year_cols
filtered_data = filtered_data[keep_cols]
print("Columns kept:", filtered_data.shape[1])

# ── STEP 4: Merge with country metadata (Region + Income Group) ──
merged_data = filtered_data.merge(real_countries, on='Country Code', how='inner')
print("Rows after merging with country metadata:", len(merged_data))

# ── STEP 5: Convert wide format to long format (years become one column) ──
df_long = merged_data.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name',
             'Indicator Code', 'Region', 'Income Group'],
    value_vars=year_cols,
    var_name='Year',
    value_name='Value'
)
df_long['Year'] = df_long['Year'].astype(int)
df_long = df_long.dropna(subset=['Value'])  # remove rows with no data
print("Final long format rows:", len(df_long))
print("\nSample of cleaned data:")
print(df_long.head(10).to_string())
````

````python
# Chart - 5 visualization code
# Bivariate Analysis - Female vs Male literacy rate comparison by region (2015)

# Filter female and male literacy for 2015
female = df_long[
    (df_long['Indicator Code'] == 'SE.ADT.LITR.FE.ZS') &
    (df_long['Year'] == 2015)
][['Country Name', 'Region', 'Value']].rename(columns={'Value': 'Female'})

male = df_long[
    (df_long['Indicator Code'] == 'SE.ADT.LITR.MA.ZS') &
    (df_long['Year'] == 2015)
][['Country Name', 'Region', 'Value']].rename(columns={'Value': 'Male'})

# Merge female and male together
gender = female.merge(male, on=['Country Name', 'Region'])
gender['Gender Gap'] = gender['Male'] - gender['Female']

# Average by region
region_gender = gender.groupby('Region')[['Female', 'Male']].mean().reset_index()
region_gender = region_gender.melt(id_vars='Region',
                                    var_name='Gender',
                                    value_name='Literacy Rate')

plt.figure(figsize=(14, 7))
sns.barplot(data=region_gender, x='Region', y='Literacy Rate',
            hue='Gender', palette=['coral', 'steelblue'])

plt.title('Female vs Male Adult Literacy Rate by Region (2015)',
          fontsize=15, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Literacy Rate (%)', fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.legend(title='Gender', fontsize=11)
plt.tight_layout()
plt.show()

print("Average Gender Gap (Male - Female) by Region:")
gap = gender.groupby('Region')['Gender Gap'].mean().sort_values(ascending=False)
print(gap.round(2).to_string())
````

````python
# Chart - 6 visualization code
# Bivariate Analysis - Female vs Male literacy rate trend over time (2000-2015)

# Filter female literacy trend
female_trend = df_long[
    df_long['Indicator Code'] == 'SE.ADT.LITR.FE.ZS'
].groupby('Year')['Value'].mean().reset_index()
female_trend['Gender'] = 'Female'

# Filter male literacy trend
male_trend = df_long[
    df_long['Indicator Code'] == 'SE.ADT.LITR.MA.ZS'
].groupby('Year')['Value'].mean().reset_index()
male_trend['Gender'] = 'Male'

# Combine both
trend = pd.concat([female_trend, male_trend])

plt.figure(figsize=(14, 6))
sns.lineplot(data=trend, x='Year', y='Value',
             hue='Gender', palette=['coral', 'steelblue'],
             linewidth=2.5, marker='o')

plt.title('Global Female vs Male Literacy Rate Trend (2000-2015)',
          fontsize=15, fontweight='bold')
plt.xlabel('Year', fontsize=13)
plt.ylabel('Average Literacy Rate (%)', fontsize=13)
plt.xticks(range(2000, 2016, 1), rotation=45)
plt.legend(title='Gender', fontsize=11)
plt.tight_layout()
plt.show()

# Print actual values
print("Average Female vs Male Literacy Rate by Year:")
pivot = trend.pivot(index='Year', columns='Gender', values='Value')
pivot['Gap (M-F)'] = pivot['Male'] - pivot['Female']
print(pivot.round(2).to_string())
````

````python
# Chart - 7 visualization code
# Bivariate Analysis - Education Spending vs Literacy Rate (Scatter Plot)

# Filter literacy rate (both sexes) for 2010
literacy = df_long[
    (df_long['Indicator Code'] == 'SE.ADT.LITR.ZS') &
    (df_long['Year'] == 2010)
][['Country Name', 'Region', 'Value']].rename(columns={'Value': 'Literacy Rate'})

# Filter education spending for 2010
spending = df_long[
    (df_long['Indicator Code'] == 'SE.XPD.TOTL.GD.ZS') &
    (df_long['Year'] == 2010)
][['Country Name', 'Value']].rename(columns={'Value': 'Spending'})

# Merge both
scatter_data = literacy.merge(spending, on='Country Name').dropna()

plt.figure(figsize=(12, 7))
sns.scatterplot(data=scatter_data, x='Spending', y='Literacy Rate',
                hue='Region', s=100, alpha=0.8)

# Add trend line
z = np.polyfit(scatter_data['Spending'], scatter_data['Literacy Rate'], 1)
p = np.poly1d(z)
x_line = np.linspace(scatter_data['Spending'].min(),
                     scatter_data['Spending'].max(), 100)
plt.plot(x_line, p(x_line), 'r--', linewidth=2, label='Trend Line')

plt.title('Government Education Spending vs Adult Literacy Rate (2010)',
          fontsize=14, fontweight='bold')
plt.xlabel('Education Spending (% of GDP)', fontsize=13)
plt.ylabel('Adult Literacy Rate (%)', fontsize=13)
plt.legend(title='Region', fontsize=10, bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# Print correlation
corr = scatter_data['Spending'].corr(scatter_data['Literacy Rate'])
print(f"Correlation between Education Spending and Literacy Rate: {corr:.2f}")
print(f"Number of countries in analysis: {len(scatter_data)}")
print("\nTop 5 highest spending countries:")
print(scatter_data.nlargest(5,'Spending')[['Country Name','Spending','Literacy Rate']].to_string())
print("\nBottom 5 lowest spending countries:")
print(scatter_data.nsmallest(5,'Spending')[['Country Name','Spending','Literacy Rate']].to_string())
````

##### 1. Why did you pick the specific chart?

A scatter plot is basically to measure the distribution of the two values - in this case, adult literacy rates and education spending. Each dot represents one country. If spending and literacy are strongly correlated then graph would form a diagonal pattern. The red line makes the overall direction of the relationaship immediately visible, and coloring dots by region adds a third dimension showing whether regional pattern exist or not.

The trend line is nearly flat, confirming the weak relationship.
If we take a look at Timor - Leste spends the most at 10.39% of GDP yet has only 58.3% literacy rate, while Sri Lanka spends just 1.72% of GDP bus has achieves 91.18% literacy. Similarly, Central African Republic spends the least (1.20%) and has very low literacy (36.75%)

````python
# Chart - 12 visualization code
# Bivariate Analysis - Top 15 countries with largest gender literacy gap (2010)

# Filter female literacy
female_lit = df_long[
    (df_long['Indicator Code'] == 'SE.ADT.LITR.FE.ZS') &
    (df_long['Year'] == 2010)
][['Country Name', 'Region', 'Value']].rename(columns={'Value': 'Female'})

# Filter male literacy
male_lit = df_long[
    (df_long['Indicator Code'] == 'SE.ADT.LITR.MA.ZS') &
    (df_long['Year'] == 2010)
][['Country Name', 'Region', 'Value']].rename(columns={'Value': 'Male'})

# Merge and calculate gap
gender_gap = female_lit.merge(male_lit, on=['Country Name', 'Region'])
gender_gap['Gap'] = gender_gap['Male'] - gender_gap['Female']
gender_gap = gender_gap.dropna(subset=['Gap'])

# Get top 15 largest gaps
top15_gap = gender_gap.nlargest(15, 'Gap')

plt.figure(figsize=(14, 8))
sns.barplot(data=top15_gap,
            x='Gap',
            y='Country Name',
            hue='Region',
            dodge=False,
            palette='Set1')

plt.title('Top 15 Countries with Largest Gender Literacy Gap (Male - Female) 2010',
          fontsize=13, fontweight='bold')
plt.xlabel('Gender Literacy Gap (Percentage Points)', fontsize=13)
plt.ylabel('Country', fontsize=13)
plt.legend(title='Region', fontsize=10,
           bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

print("Top 15 countries with largest gender literacy gap (2010):")
print(top15_gap[['Country Name', 'Region', 'Female', 'Male', 'Gap']].to_string())
print(f"\nLargest gap: {top15_gap.iloc[0]['Country Name']} — {top15_gap.iloc[0]['Gap']:.1f} points")
print(f"Global average gap: {gender_gap['Gap'].mean():.1f} points")
````
